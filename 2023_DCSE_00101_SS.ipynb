## Question 2: Matrices and Numerical Python

Construct a matrix below ; 

$$\left[\begin{array}{ccc} 1 & 0.2 & 0.5 \\ 0.2 & 1 & 0.8 \\ 0.5 & 0.8 & 1 \end{array}\right]$$

as a NumPy array.


(a). Compute an Inversion of the matrix (`Transpose`) 

(b). Computer the (`determinant`) of the matrix if possible 

```python
# Supply the code for the Question 2 (5 marks)
```

```python
import numpy as np
```

```python
matrix = np.array([[1,0.2,0.5],[0.2,1,0.8],[0.5,0.8,1]])
print (matrix)
```

```python
#matrix transpose
matrix_transpose = np.transpose(matrix)
print(matrix)
print(matrix_transpose)
```

```python
#matrix determinant
matrix = np.array([[1,0.2,0.5],[0.2,1,0.8],[0.5,0.8,1]])
matrix_determinant = np.linalg.det(matrix)
print(matrix_determinant)
```
